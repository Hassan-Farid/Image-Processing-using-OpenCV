```python
import cv2
import numpy as np
```

```python
#Loading the image
image = cv2.imread('images\image2.jpg')
```

```python
#Creating the binarized image 
retval, binary = cv2.threshold(image, 85, 255, cv2.THRESH_BINARY)
```

```python
#Displaying the actual image
cv2.imshow('Image Window', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Displaying the binarized image
cv2.imshow('Binarized Image', binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Creating the binarized inverted image
retval, binary_inv = cv2.threshold(image, 85, 255, cv2.THRESH_BINARY_INV)
```

```python
#Displaying the binarized inverted image
cv2.imshow('Binarized Inverted Image', binary_inv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Creating the grayscale image
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#Displaying the gray scale image
cv2.imshow('GrayScale Image', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Creating truncated image
retval, trunc_img = cv2.threshold(image, 85, 255, cv2.THRESH_TRUNC)
```

```python
#Displaying the trucated image
cv2.imshow('Trucated Image', trunc_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Creating the Zero Threshold image
retval, to_zero_img = cv2.threshold(image, 85, 255, cv2.THRESH_TOZERO)
```

```python
#Displaying the Zero Threshold image
cv2.imshow('Zero Image', to_zero_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Creating the Zero Inverted Threshold image
retval, zero_inv_img = cv2.threshold(image, 85, 255, cv2.THRESH_TOZERO_INV)
```

```python
#Displaying the Zero Inverted image
cv2.imshow('Zero Inverted Image', zero_inv_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Now using matplotlib lets check the difference between the obtained images using subplots
import matplotlib.pyplot as plt

titles = ['Image Window', 'GrayScale Image', 'Binary Image', 'Binary Inverted Image', 'Trucated Image', 'Zero Image', 'Zero Inverted Image']
images = [image, gray, binary, binary_inv, trunc_img, to_zero_img, zero_inv_img]

for i in range(6):
    plt.subplot(2,3,i+1) #Creates the subplot at the (i+1)th location in a 2x3 table of plots
    plt.imshow(images[i]) #To display the images 
    plt.title(titles[i]) #To display the titles
    plt.xticks([]) #To avoid any xticks in the plot
    plt.yticks([]) #To avoid any yticks in the plot
    
plt.show() #Displays the image on the screen
```

```python
#Adaptive Thresholding
#Some times in a given image, one global value for thresholding is not enough, for those cases we use adaptive thresholding e.g. a book with different light conditions on various areas

image = cv2.imread('images\image3.jpg', 0) #Loading image in grayscale
```

```python
#Displaying the image
cv2.imshow('Image Window', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#In this image case global thresholding would not give a good result
#Creating the global threshold image (Binarized form)
retval, binary_img = cv2.threshold(image, 55, 255, cv2.THRESH_BINARY)
```

```python
#Displaying the threshold image
cv2.imshow('Binary Image', binary_img) #Either one side will be black and other white or vice versa
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Thus, we use adaptive thresholding to get a clear output in such cases
#Adaptive Mean Thresholding: This gives threshold value by subtracting constant passed to it from the mean of the values in the block specified
#Creating adaptive mean threshold image
adp_mean_img = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
#cv2.adaptiveThreshold(image_obj, changed_value, adpative_threshold_type, threshold_type, block_size, constant_value)
```

```python
#Displaying the adaptive mean value image
cv2.imshow('Mean Value Image', adp_mean_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Adaptive Gaussian Thresholding: This gives threshold value by subtracting constant passed to it from the gaussian sum of the values in the block specified
#Creating adaptive gaussian sum threshold image
adp_gauss_img = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11 ,2)
```

```python
#Displaying the adaptive gaussian sum image
cv2.imshow('Gaussian Sum Image', adp_gauss_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Now displaying all images to see the difference between them
images = [image, binary_img, adp_mean_img, adp_gauss_img]
titles = ['Image', 'Binary Image', 'Mean Value Image', 'Gaussian Sum Image']

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i], 'gray') #Displays images in grayscale in format
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
    
plt.show()
```

```python
#There is another method for binarization apart from global thresholding methods called Otsu Binarization
#This method does not take a particular value for thresholding but calculates it automatically
#The method is only applicable for bimodal images

#Loading a bimodal image
image = cv2.imread('images\image4.jpg', 0)
```

```python
#Creating otsu binarized object
retval , otsu_bin_img = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
#For otsu binarization, we add otsu threshold to the global threshold values
```

```python
cv2.imshow('Image' , otsu_bin_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Seeing difference between the images
images = [image, otsu_bin_img]
titles = ['Image', 'Otsu Binary Image']

for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(images[i], 'gray') #Displays images in grayscale format
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
    
plt.show()
```
