```python
import cv2
import numpy as np
```

```python
#Reading/Loading an already saved image

image = cv2.imread('images\image1.jpg', cv2.IMREAD_COLOR) #The image path and state of loaded image are passed as arguments
```

```python
#Displaying image on the screen

cv2.imshow('Image Window',image) #The title of the image window and image to be displayed are passed as arguments
cv2.waitKey(0) #The time in milliseconds for which the image will be displayed

cv2.destroyAllWindows() #Destroys all windows opened by OpenCV

#The above code would display the image but outside of the notebook
```

```python
#Displaying image on the screen as output of teh command
#Since we are using jupyter, we would like to show the image output on the screen here
#To perform that, we are going to use the plotting library matplotlib

import matplotlib.pyplot as plt

plt.imshow(image)
plt.title('Image Window')
```

```python
#In case of reading, we mostly use grayscale format, because it is easier to apply processes on
image = cv2.imread('images/image1.jpg',cv2.IMREAD_GRAYSCALE)

#Now displaying the image
plt.imshow(image, cmap = 'gray')
plt.title('Image Window')
```

```python
#In case of reading, we can also use the image unchanged
image = cv2.imread('images/image1.jpg',cv2.IMREAD_UNCHANGED)

#Now displaying the image
plt.imshow(image)
plt.title('Image Window')
```

```python
#We can use numbers as well instead of writing the IMREAD_ arguments, we have:
#IMREAD_COLOR: 1
#IMREAD_GRAYSCALE: 0
#IMREAD_UNCHANGED: -1

image = cv2.imread('images/image1.jpg',0)

#Now displaying the image
plt.imshow(image, cmap='gray')
plt.title('Image Window')
```
